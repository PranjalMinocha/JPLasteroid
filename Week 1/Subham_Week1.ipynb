```python
q1=df.a.quantile(0.25)
q3=df.a.quantile(0.75)
iqr=q3-q1
uc=q1+iqr*1.5
lc=q1-iqr*1.5
print(f'{uc} and {lc}')
```

```python
sns.relplot(x="a", y="diameter", kind="line", data=df)
plt.xlim([0,3.69])
```

by looking at the relation between a and diameter it is clearly visible that outliers below the lower cuttoff is not related with diameter so they can be removed

Similarly for each feature i have checked for their relation with diameter above upper cuttoff and below lower cuttoff and then decided on what should be cliped and in some case i have put some upper cutoff simply on basis of graph relation over which the parameter is not related with diameter.
But in few case like in q the upper value were highly related with diameter so i haven'nt cliped the data

```python
df.a.clip(upper=uc,lower=lc,inplace=True)
```

```python
sns.relplot(x="a", y="diameter", kind="line", data=df)
```

```python
df.e.clip(upper=0.5,inplace=True)
```

```python
q1=df.i.quantile(0.25)
q3=df.i.quantile(0.75)
iqr=q3-q1
uc=q1+iqr*1.5
df.i.clip(upper=uc,inplace=True)
```

```python
sns.relplot(x="q", y="diameter", kind="line", data=df)
plt.xlim([0,3.66])
```

```python
df.q.clip(lower=1.14,inplace=True)
```

```python
sns.relplot(x="ad", y="diameter", kind="line", data=df)
plt.xlim([0,10])
```

```python
df.ad.clip(upper=6,inplace=True)
```

```python
df.albedo.clip(upper= 0.390500,inplace=True)
```

```python
df.n.clip(upper= 0.338023,inplace=True)
```

```python
df.H.clip(upper= 18,inplace=True)
```

```python
df['albedo'].fillna(value=df['albedo'].mean(),inplace=True)
df['H'].fillna(value=df['H'].mean(),inplace=True)
df['data_arc'].fillna(value=df['data_arc'].mean(),inplace=True)
```
