```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Asteroid_Modified.csv',low_memory=False)
```

# Standardization of Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled = pd.DataFrame(data = df_scaled, columns = df.columns.values.tolist())
```

```python
df_scaled
```

# Test-Train Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
input_features = df_scaled.columns.values.tolist()
input_features.remove('diameter')
```

```python
X = df_scaled[[i for i in input_features]]
```

```python
y = df_scaled[['diameter']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , random_state=42, test_size=0.2, shuffle=True)
```

# Model Training

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, median_absolute_error, explained_variance_score, r2_score
```

```python
def prediction_metrics(prediction):
    print("Mean Absolute Error :     ",mean_absolute_error(y_test, prediction))
    print("Mean Squared Error :      ",mean_squared_error(y_test, prediction))
    print("Median Absolute Error :   ",median_absolute_error(y_test, prediction))
    print("Explained Variance Score :",explained_variance_score(y_test, prediction))
    print("r2-Score :                ",r2_score(y_test, prediction))
```

```python
def cv_display_scores(scores):
    print("Scores: ", scores)
    print("Mean:              ", scores.mean())
    print("Standard deviation:", scores.std())
```

## Ensembles

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import HistGradientBoostingRegressor

from sklearn.svm import SVR

from xgboost import XGBRegressor

from lightgbm import LGBMRegressor
```

### Random Forest Regressor

```python
forest_reg = RandomForestRegressor()
forest_scores = cross_val_score(forest_reg, X, y.values.ravel(), scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
cv_display_scores(forest_rmse_scores)
```

```python
forest_reg.fit(X_train, y_train.values.ravel())
forest_reg_pred = forest_reg.predict(X_test)
prediction_metrics(forest_reg_pred)
```

### Extra Trees Regressor

```python
extratree_reg = ExtraTreesRegressor(n_estimators=10, max_depth=None, min_samples_split=2, random_state=0)
extratree_scores = cross_val_score(extratree_reg, X, y, scoring="neg_mean_squared_error", cv=10)
extratree_rmse_scores = np.sqrt(-extratree_scores)
cv_display_scores(extratree_rmse_scores)
```

```python
extratree_reg.fit(X_train, y_train.values.ravel())
extratree_reg_pred = extratree_reg.predict(X_test)
prediction_metrics(extratree_reg_pred)
```

### ADA Boost Regressor

```python
ada_reg = AdaBoostRegressor(random_state=0, n_estimators=100)
ada_scores = cross_val_score(ada_reg, X, y, scoring="neg_mean_squared_error", cv=10)
ada_rmse_scores = np.sqrt(-ada_scores)
cv_display_scores(ada_rmse_scores)
```

```python
ada_reg.fit(X_train, y_train.values.ravel())
ada_reg_pred = ada_reg.predict(X_test)
prediction_metrics(ada_reg_pred)
```

### Gradient Boost Regressor

```python
gb_reg = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=1, random_state=0, loss='squared_error')
gb_scores = cross_val_score(gb_reg, X, y, scoring="neg_mean_squared_error", cv=10)
gb_rmse_scores = np.sqrt(-gb_scores)
cv_display_scores(gb_rmse_scores)
```

```python
gb_reg.fit(X_train, y_train.values.ravel())
gb_reg_pred = gb_reg.predict(X_test)
prediction_metrics(gb_reg_pred)
```
