**a - SEMI MAJOR AXIS**

The semi-major axis, a, is half of the longest diameter of an ellipse. Together with the semi-minor axis , b, and eccentricity, e, it forms a set of related values that completely describe the shape of an ellipse:

b2 = a2(1-e2)

In cartesian coordinates (x,y), an ellipse is the solution of:

$ \left(\frac xa\right)^2 + \left(\frac yb\right)^2 = 1 $

<strong>Sequential Regressor</strong>

```python
import tensorflow as tf

from tensorflow import keras
from keras.datasets import mnist
from keras import layers

from keras.layers import Dense
from keras.layers import Dropout
```

```python
def build_model():
    model = keras.Sequential([
        layers.Dense(64, activation=tf.nn.relu, input_shape=[28]),
        layers.Dense(64, activation=tf.nn.relu),
        layers.Dense(1)
    ])
    
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss='mse',
                 optimizer=optimizer,
                 metrics=['mse'])
    return model

```

```python
model = build_model()
```

```python
model.summary()
```

```python
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')
        

EPOCHS = 100

history = model.fit(
    X_train, Y_train,
    epochs=EPOCHS, validation_data=(X_test, Y_test),verbose=0,
    callbacks=[PrintDot()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
score = model.evaluate(X_test, Y_test, verbose=0)

r2 = r2_score(model.predict(X_test), Y_test)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print(r2)
```
